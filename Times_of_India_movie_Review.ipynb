```python
import sys
import time
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://timesofindia.indiatimes.com/entertainment/chennai/movie-showtimes/40"
try:

    page=requests.get(url)

except Exception as e:
    error_type, error_obj, error_info = sys.exc_info()
    print ('ERROR FOR LINK:',url)
    print (error_type, 'Line:', error_info.tb_lineno)

time.sleep(2)
soup=BeautifulSoup(page.text,'html.parser')
links=soup.find_all('div',{'class':"movie_info_box clearfix"})
```

```python
links
```

```python
movie =0
movie_list={}
for div in links:
    title = div.find('h2').text
    print(title)
    stars = div.find('p').text
    print(stars)
    ratings_tag = div.find('span',{'class':'star_count'})
    ratings = ratings_tag.text if ratings_tag else "N/A"
    genre = div.find('p',{'class':'clearfix'}).text
    print(genre)
    print(ratings)
    place = div.find('h3').text
    print(place)
    timing_disabled_tag = div.find_all('span',{'class':'timing_box disabled'})
    for i in timing_disabled_tag:
      timing_disabled = i.text if timing_disabled_tag else "N/A"
      print("timing_disabled - ",timing_disabled)
    timing_available_tag = div.find('span',{'class':'timing_box'})
    for i in timing_available_tag:
      timing_available = i.text if timing_available_tag else "N/A"
      print("timing_available - ",timing_available)
    movie += 1
    movie_list[movie] = [title,stars,ratings,genre,place,timing_disabled,timing_available]
    print("\n")
```

```python
print("Total movies",movie)
movie_list_df = pd.DataFrame.from_dict(movie_list,orient='index',columns=['title','stars','ratings','genre','place','timing_disabled','timing_available'])
```
